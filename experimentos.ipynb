```python
def start_chat(data, tipo, token):
    """
    Realiza uma requisição POST para a rota /start_chat.

    Parâmetros:
        - data (dict): O payload a ser enviado no corpo da requisição.
        - tipo (str): Tipo de ferramenta ('COMPLETO', 'RESUMO' ou vazio).
        - token (str): Token de autenticação.

    Retorna:
        - dict: Resposta do servidor, se a requisição for bem-sucedida.
    """
    url = f'http://127.0.0.1:5000/start_chat?tipo={tipo}'
    headers = {
        'Authorization': f'Bearer {token}',
        'Content-Type': 'application/json',
        'User-Agent': 'insomnia/10.3.0',
    }

    try:
        response = requests.post(url, headers=headers, data=json.dumps(data))
        response.raise_for_status()  # Levanta uma exceção para status HTTP 4xx/5xx
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}

def continue_chat(chat_id, pergunta, tipo, token):
    """
    Realiza uma requisição POST para a rota /continue_chat.

    Parâmetros:
        - data (dict): O payload a ser enviado no corpo da requisição.
        - tipo (str): Tipo de ferramenta ('COMPLETO', 'RESUMO' ou vazio).
        - token (str): Token de autenticação.

    Retorna:
        - dict: Resposta do servidor, se a requisição for bem-sucedida.
    """
    url = f'http://127.0.0.1:5000/continue_chat?tipo={tipo}'
    headers = {
        'Authorization': f'Bearer {token}',
        'Content-Type': 'application/json',
        'User-Agent': 'insomnia/10.3.0',
    }

    try:
        data = {
                "chat_store_key":chat_id,
	            "new_message":pergunta
                }
        response = requests.post(url, headers=headers, data=json.dumps(data))
        response.raise_for_status()  # Levanta uma exceção para status HTTP 4xx/5xx
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}
```
